安裝好 OpenCV 之後分別顯示同一張圖片但個別是彩色 RGB 格式跟灰階 Grayscale 格式的圖

可以選擇電腦中任意圖片（建議選擇 .jpg, .jpeg, .png)

或是電腦視覺領域中知名的 [Lena](http://optipng.sourceforge.net/pngtech/img/lena.png) 圖皆可

```python
import cv2
from  matplotlib import pyplot as plt
%matplotlib inline

def plt_cvimg(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.show()
```

matplotlib 顏色通道為 RGB
opencv 顏色通道為 BGR
使用matplotlib顯示之前需要轉換

```python
img_path = 'data/Lenna.png'

# 載入圖片 - 彩色圖
img = cv2.imread(img_path, cv2.IMREAD_COLOR)

# 載入圖片 - 灰階圖
img_gray = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
# 將灰階圖轉為三通道
img_gray = cv2.cvtColor(img_gray, cv2.COLOR_GRAY2BGR)
```

```python
plt_cvimg(img)
plt_cvimg(img_gray)
```
